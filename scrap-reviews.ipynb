```python
#importing all libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
```

```python
driver=webdriver.Chrome(r"chromedriver.exe") 
time.sleep(4)

url = "https://www.flipkart.com/"
driver.get(url)#entering into flipkart.com in googlechrome
time.sleep(3)
```

```python
close=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close.click()#close login 
```

```python
driver.find_element_by_xpath("//input[@class='_3704LK']").send_keys("laptop")#enter laptop in search bar
time.sleep(2)
```

```python
search=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")#searching laptop in the site
search.click()
```

```python
urls=[]#appending urls of product
for i in range(10):
    for j in driver.find_elements_by_xpath("//div[@class='_2kHMtA']/a"):
        urls.append(j.get_attribute("href"))
    try:
        driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]").click()
    except:
        driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
    time.sleep(4)
```

```python
ratings=[]
full_reviews=[]
for k in urls:
    driver.get(k)
    time.sleep(3)
    try:
        driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click() # clcik on see all reviews
        time.sleep(2)
        # fetching first pages reviews
        for a in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
            full_reviews.append(a.text)
        for b in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
            ratings.append(b.text)
        time.sleep(1)
        #clciking on second page for reviews
        driver.find_element_by_xpath("//a[@class='_1LKTO3']").click()
        time.sleep(2)
        # running loop till the end of pages of all reviews
        while True:
            time.sleep(2)
            for i in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
                full_reviews.append(i.text)
            for j in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
                ratings.append(j.text)
            try:
                driver.find_element_by_xpath("//a[@class='_1LKTO3'][2]/span").click()
                time.sleep(1)
            except:
                break
    except:
        print("no reviews")
```

```python
df=pd.DataFrame({})#making dataframe
df['Full_Reviews']=full_reviews
df['Ratings']=ratings
df
```

```python
df.to_csv("rev.csv")#store it in csv file in this directory
```
